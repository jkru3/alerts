```python
import re

def extract_company_names():
    # Read the content from file
    with open('paste.txt', 'r') as file:
        content = file.read()
    
    # Use regex to extract company names
    pattern = r"name: '([^']+)'"
    companies = re.findall(pattern, content)
    
    # Create formatted checklist
    checklist = ""
    for company in sorted(set(companies)):
        checklist += f"- [ ] {company}\n"
    
    # Write to output file
    with open('company_checklist.txt', 'w') as output_file:
        output_file.write(checklist)
    
    print(f"Generated checklist with {len(set(companies))} companies.")
    print("Checklist saved to company_checklist.txt")
    
    # Also print the checklist to console
    print("\nCompany Checklist:")
    print(checklist)

if __name__ == "__main__":
    extract_company_names()
```
